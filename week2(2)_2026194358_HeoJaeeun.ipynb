```python
name = "Heo Jaeeun"
len(name)  # len (length, 문자열의 길이)
```

```python
print(name[5])  # index에서는 대괄호 [] 사용
print(name[-3]) # index는 0부터 시작
```

```python
first_name = name[4:10]  # index 4 부터 index 9 까지
print(first_name)
last_name = name[0:3]    # index 0 부터 index 2 까지
print(last_name)

print("My name is " + first_name + '!')
```

```python
# name[start:end:step]
print(name[::2])      # 전체 index에서 2칸씩 건너뛰면서 가져오기: index 0, index 2, index 4, ...
print(name[2:10:3])   # index 2부터 index 9까지 3칸씩 건너뛰면서 가져오기: index 2, index 5, index 8
```

```python
# find 함수
print(name.find('ee'))     # ee가 시작된 index 6 출력
print(name.find('abcd'))   # 찾을 수 없으면 -1 출력
```

```python
# upper, lower 함수
A = "My name is Jaeeun."
print("Before: " + A)
B = A.upper()
print("UPPER : " + B)
C = A.lower()
print("lower : " + C)
```

```python
# replace 함수
A = "I like apples!"
print(A)
B = A.replace("apples", "bananas")
print(B)
C = B.replace("like", "HATE")
print(C)
```
