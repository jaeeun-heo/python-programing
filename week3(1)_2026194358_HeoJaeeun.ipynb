tuple: a collection that groups different types of data(string, integer, or float) into one unit

```python
tuple_1 = ("apple", '10', '1.2')
tuple_1
```

```python
print(tuple_1[0])
type(tuple_1[0])
```

```python
tuple_2 = ("banana", '100', '10.2')
tuple_3 = tuple_1 + tuple_2
tuple_3
```

```python
print(type(tuple_3))
print(len(tuple_3))
```

```python
rating = (2, 0, 2, 6, 1, 9, 4, 3, 5, 8)
print(type(rating))
print("==========")

rating_sorted = sorted(rating)  # sort는 정수(integer)와 실수(float)만 가능
print(rating_sorted)
print(type(rating_sorted))
```

```python
nested_tuple = (1, 2, ("pop", "rock"), (3, 4), ("disco", (1, 2)))
print("nested_tuple:", nested_tuple)
print("===" * 25)
print("element 0 of tuple:", nested_tuple[0], "\t\t\t|  data type:", type(nested_tuple[0]))
print("element 1 of tuple:", nested_tuple[1], "\t\t\t|  data type:", type(nested_tuple[1]))
print("element 2 of tuple:", nested_tuple[2], "\t|  data type:", type(nested_tuple[2]))
print("element 3 of tuple:", nested_tuple[3], "\t\t|  data type:", type(nested_tuple[3]))
print("element 4 of tuple:", nested_tuple[4], "\t|  data type:", type(nested_tuple[4]))
print("===" * 25)
print("element 2, 0 of tuple:", nested_tuple[2][0], "\t\t|  data type:", type(nested_tuple[2][0]))
print("element 2, 0, 0 of tuple:", nested_tuple[2][0][0])
print("element 2, 0, 1 of tuple:", nested_tuple[2][0][1])
print("element 2, 0, 2 of tuple:", nested_tuple[2][0][2])
print("element 4, 1 of tuple:", nested_tuple[4][1], "\t\t|  data type:", type(nested_tuple[4][1]))
print("element 4, 1, 0 of tuple:", nested_tuple[4][1][0])
print("element 4, 1, 1 of tuple:", nested_tuple[4][1][1])
```

### **Quiz on Tuples**

#### Consider the following tuple:

```python
genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", "R&B", "progressive rock", "disco")
genres_tuple
```

#### Find the length of the tuple, *genres_tuple*:

```python
len(genres_tuple)
```

#### Access the element, with respect to index 3:

```python
genres_tuple[3]
```

#### Use slicing to obtain indexes 3, 4 and 5:

```python
genres_tuple[3:6]
```

#### Find the first two elements of the tuple *genres_tuple*:

```python
genres_tuple[0:2]
```

#### Find the first index of *"disco"*:

```python
genres_tuple.index("disco")   # tuple에서는 find 대신 index 사용
```

#### Generate a sorted List from the Tuple *C_tuple=(-5, 1, -3)*:

```python
C_tuple = (-5, 1, -3)
C_tuple_sorted = sorted(C_tuple)
C_tuple_sorted
```
