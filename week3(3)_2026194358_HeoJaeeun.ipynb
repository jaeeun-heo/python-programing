## **Dictionaries : week 3-3**

```python
dict = {"key 1": 1, "Key 2": "2", "key 3": (3, 3, 3), "key 4": [4, 4, 4], 5.5: 55, (6, 6): 66}
dict

# dictionaries는 중괄호로 표현 {}
# key 1: 1 (integer)
# key 2: "2" (string)
# key 3: (3, 3, 3) (tuple)
# key 4: [4, 4, 4] (list, key로는 사용할 수 없음)
```

```python
fruit = {"apple": "red", "banana": "yellow", "kiwi": "green", "grape": "purple"}
print(fruit['apple'])
print("keys:", fruit.keys())
print("values:", fruit.values())
```

```python
# add
fruit['bear'] = 'brown'
print(fruit)

# delete
del(fruit['bear'])
print(fruit)
```

```python
# 찾기
print('kiwi' in fruit)      # bool 값으로 출력
print('purple' in fruit)    # value는 false로 출력
```
