```python
# copy
A = ["Apple", "Angel", "About"]
B = A
A[0] = "Banana"
print("A:", A)
print("B:", B)   # A의 변경 사항이 B에도 적용, B의 변경 사항도 A에 적용됨
```

```python
# clone
A = ["Apple", "Angel", "About"]
B = A[:]
A[0] = "Banana"
print("A:", A)
print("B:", B)   # A의 변경 사항이 B에 적용되지 않음
```
