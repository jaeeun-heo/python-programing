## **Functions**

```python
def add(x):   # def; define / add: function / x: parameter (or argument)
    y = x + 1
    return(x, y)   # 출력 형태

add(1)   # 함수 호출
```

```python
def mult(x, y):
    z = x * y
    return(z)

mult(2, 6)
```

```python
def square(a):
    c = 3
    b = a * 2 + c   # a, b, c: local variable, 함수 안에서 정의된 값
    return(b)

square(4)
```

```python
def con(m, n):
    return(m + n)

con("HY", "MOT")
```

```python
def equational(a, b):
    c = a * a + b * b + 2 * a * b
    if (c < 25):
        d = "small"
    elif (c > 25) and (c < 50):
        d = "medium"
    else:
        d = "big"
    return(c, d)

print(equational(1, 2))
print(equational(3, 4))
print(equational(5, 6))
```

```python
def HY():
    global x
    x = 2026194358  # global variable
    return(x)

HY()       # 호출 후
print(x)   # 출력 가능
```
