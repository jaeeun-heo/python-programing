## **Classes and Objects**

```python
# import the library, 그림 그리기
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Rectangle
class Rectangle(object):
    def __init__(self, width, height, color):   # 불러오기
        self.width = width
        self.height = height
        self.color = color
    def drawRectangle(self):   # 함수 정의
        plt.gca().add_patch(plt.Rectangle((0,0), self.width, self.height, fc=self.color))
        plt.axis("scaled")
        plt.show()
    def add_width(self, w):   # 폭 변경하기
        self.width = self.width + w
        return(self.width)
    def add_height(self, h):   # 높이 변경하기
        self.height = self.height + h
        return(self.height)
```

```python
spah = Rectangle(3, 5, 'skyblue')
spah.drawRectangle()
```

```python
End_spah = Rectangle(2, 2, 'brown')
End_spah.add_width(1)   # width + 1
End_spah.add_width(2)   # width + 1 + 2
End_spah.add_height(2)  # height + 2
End_spah.add_height(-1) # height + 2 - 1
End_spah.drawRectangle()   # draw
```

```python
# Circle
class Circle(object):
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0,0), radius=self.radius, fc=self.color))
        plt.axis("scaled")
        plt.show()
    def add_radius(self, r):   # 반지름 변경하기
        self.radius = self.radius + r
        return(self.radius)
```

```python
ehdrmfl = Circle(2.5, 'purple')
ehdrmfl.drawCircle()
```

```python
big_ehdrmfl = Circle(5, 'green')
big_ehdrmfl.add_radius(3)   # 반지름 + 3
big_ehdrmfl.drawCircle()    # 그리기
```
