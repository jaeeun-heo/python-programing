## **Loops in Python**

```python
ID = [2, 0, 2, 6, 1, 9, 4, 3, 5, 8]

for i in ID:
    print(i)
```

```python
i = 0   # index 0부터

while(i != 5):   # i가 5가 되기 전까지 반복
    I = ID[i]    # 리스트 ID에서 i번째 요소를 추출하여 I에 입력
    i = i + 1    # i는 (0부터) 1씩 증가
    print(I)     # I를 출력
```

### **Quiz on Loops**

#### Write a *for* loop the prints out all the element between -5 and 5 using the range function.

```python
for i in range(-5, 6):   # -5부터 6이 되기 전까지
    print (i)
```

#### Print the elements of the following list: *Genres = ["rock", "R&B", "soundtrack", "R&B", "soul", "pop"]*

```python
Genres = ["rock", "R&B", "soundtrack", "R&B", "soul", "pop"]

for i in Genres:
    print (i)
```

#### Write a for loop that prints out the following list: Squares = ["red", "yellow", "green", "purple", "blue"]

```python
Squares = ["red", "yellow", "green", "purple", "blue"]

for i in Squares:
    print (i)
```

#### Write a while loop to display the values of the Rating of an album playlist stored in the list *PlayListRating*. If the score is less than 6, exit the loop. The list is given by: PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]

```python
PLR = [10, 9.5, 10, 8, 7.5, 5, 10, 10]

i = 0

while(PLR[i] >= 6):   # 요소가 6보다 같거나 작게 되기 전까지 반복
    print(PLR[i])
    i = i + 1
```

#### Write a while loop to copy the strings 'orange' of list *squares* to the list *new_squares*. Stop and exit the loop if the value on the list is not 'orange':

```python
squares = ["orange", "orange", "purple", "blue", "orange"]
new_squares = []

i = 0

while(squares[i] == "orange"):       # 요소가 'orange'가 되지 않을 때까지 반복('orange'이면 반복)
    new_squares.append(squares[i])   # 기존 new_squares 리스트에 요소가 없으므로 append로 추가하는 과정 필요
    i = i + 1
print(new_squares)   # 루프 빠져나와 최종 결과만 출력, index 2에서 종료되어 index 4의 orange는 추가되지 못함
```
